## Topic 1: Lectura y escritura básica de archivos de texto

Python usa la función `open()` para abrir archivos. Modos comunes:
- `'r'` → lectura (default)
- `'w'` → escritura (sobrescribe)
- `'a'` → append (agrega al final)
- `'r+'` → lectura y escritura

Siempre es mejor usar `with` para cerrar automáticamente el archivo.

```python
# Demostración: Escritura y lectura básica
with open('ejemplo.txt', 'w') as archivo:
    archivo.write('Línea 1\n')
    archivo.write('Línea 2\n')
    archivo.write('Línea 3\n')

with open('ejemplo.txt', 'r') as archivo:
    contenido = archivo.read()
    print("Contenido:\n", contenido)
```

## Topic 2: Context manager `with` y modos avanzados

El bloque `with` asegura que el archivo se cierre incluso si ocurre un error.

Otros métodos útiles:
- `readline()` → lee una línea
- `readlines()` → lista de líneas
- `writelines()` → escribe lista de strings

```python
# Demostración: readlines y append
with open('ejemplo.txt', 'a') as archivo:
    archivo.write('Línea agregada\n')

with open('ejemplo.txt', 'r') as archivo:
    lineas = archivo.readlines()
    print("Líneas:")
    for linea in lineas:
        print(linea.strip())
```

## Topic 3: Trabajo con CSV y JSON

- **CSV**: usa módulo `csv`
- **JSON**: usa módulo `json` (serialize/deserialize datos Python ↔ JSON)

```python
# Demostración: CSV simple
import csv

datos = [['Nombre', 'Edad'], ['Ana', 28], ['Luis', 35]]

with open('personas.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(datos)

# Lectura
with open('personas.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
# Demostración: JSON
import json

persona = {"nombre": "Sofía", "edad": 31, "ciudad": "Barcelona"}

with open('persona.json', 'w') as f:
    json.dump(persona, f, indent=4)

with open('persona.json', 'r') as f:
    datos = json.load(f)
    print(datos)
```

## Topic 4: Manejo de excepciones – try / except / else / finally

Captura errores para evitar que el programa se caiga.

Mejores prácticas:
- Capturar excepciones específicas (ValueError, FileNotFoundError, etc.)
- Usar `else` (se ejecuta si NO hay error)
- `finally` siempre se ejecuta (limpieza)

```python
# Demostración: Manejo de excepciones
try:
    with open('archivo_inexistente.txt', 'r') as f:
        contenido = f.read()
except FileNotFoundError:
    print("El archivo no existe. Creando uno nuevo...")
    with open('archivo_inexistente.txt', 'w') as f:
        f.write("Archivo creado por manejo de error")
except Exception as e:
    print(f"Error inesperado: {e}")
else:
    print("Archivo leído correctamente")
finally:
    print("Operación finalizada.")
```

### Reto 1: Crear y escribir lista de tareas

Crea un archivo 'tareas.txt' y escribe 3 tareas. Guarda el nombre del archivo en `archivo_tareas`.

```python
# Tu código aquí
archivo_tareas = "tareas.txt"

with open(archivo_tareas, 'w') as f:
    f.write("Comprar leche\n")
    f.write("Llamar al banco\n")
    f.write("Estudiar Python\n")
```

```python
# Validación suave
try:
    with open(archivo_tareas, 'r') as f:
        contenido = f.read()
    assert "Estudiar Python" in contenido, "Falta al menos una tarea"
    print("¡Bien! Sigue al siguiente reto.")
except FileNotFoundError:
    print("Intenta de nuevo: No se creó el archivo.")
except AssertionError as e:
    print(f"Intenta de nuevo: {e}")
```

### Reto 2: Leer y agregar tarea nueva

Lee `archivo_tareas`, agrega una tarea nueva y cuenta líneas. Guarda conteo en `num_tareas`.

```python
# Fallback si no existe archivo_tareas
try:
    archivo_tareas
except NameError:
    archivo_tareas = "tareas.txt"
    with open(archivo_tareas, 'w') as f:
        f.write("Tarea de ejemplo\n")

# Tu código aquí
with open(archivo_tareas, 'a') as f:
    f.write("Hacer ejercicio\n")

with open(archivo_tareas, 'r') as f:
    lineas = f.readlines()
    num_tareas = len(lineas)
```

```python
# Validación suave
try:
    assert num_tareas >= 4, f"Debería haber al menos 4 tareas. Tienes {num_tareas}"
    print("¡Bien! Sigue al reto final.")
except AssertionError as e:
    print(f"Intenta de nuevo: {e}")
```

#@markdown Pista: 'a' para append, len(f.readlines()) para contar líneas

### Reto Final: Manejo seguro con excepciones

Intenta leer un archivo que puede no existir. Si falla, crea uno con mensaje. Guarda estado en `operacion_exitosa`.

```python
# Fallbacks
try:
    num_tareas
except NameError:
    num_tareas = 1

# Tu código aquí
archivo_log = "log.txt"
operacion_exitosa = False

try:
    with open(archivo_log, 'r') as f:
        contenido = f.read()
except FileNotFoundError:
    with open(archivo_log, 'w') as f:
        f.write(f"Archivo creado automáticamente. Tareas previas: {num_tareas}\n")
    operacion_exitosa = True
except Exception as e:
    print(f"Error inesperado: {e}")
else:
    operacion_exitosa = True
finally:
    print("Operación finalizada.")
```

```python
# Validación simple
try:
    assert os.path.exists(archivo_log), "El archivo log.txt debería existir"
    print("¡Éxito! Listo para el proyecto.")
except AssertionError as e:
    print(f"Revisa: {e}")
```

## Proyecto: Gestor de tareas persistente

Crea un gestor simple que:
- Permita agregar tareas a un archivo
- Lea y muestre todas las tareas
- Maneje errores (archivo no existe, permisos, etc.)
- Use funciones para modularidad

```python
# Tu código aquí - Proyecto: Gestor de tareas
ARCHIVO_TAREAS = "mis_tareas.txt"

def agregar_tarea(tarea):
    try:
        with open(ARCHIVO_TAREAS, 'a') as f:
            f.write(tarea + "\n")
        return True
    except Exception as e:
        print(f"Error al agregar: {e}")
        return False

def mostrar_tareas():
    try:
        with open(ARCHIVO_TAREAS, 'r') as f:
            tareas = f.readlines()
        if tareas:
            print("Tareas pendientes:")
            for i, t in enumerate(tareas, 1):
                print(f"{i}. {t.strip()}")
        else:
            print("No hay tareas.")
    except FileNotFoundError:
        print("Archivo no encontrado. No hay tareas aún.")
    except Exception as e:
        print(f"Error al leer: {e}")

# Pruebas
agregar_tarea("Comprar pan")
agregar_tarea("Estudiar excepciones")
mostrar_tareas()
```

```python
# Validación simple del proyecto
try:
    with open(ARCHIVO_TAREAS, 'r') as f:
        assert len(f.readlines()) >= 2, "Deberías haber agregado al menos 2 tareas"
    print("Proyecto completado. ¡Buen trabajo!")
except Exception as e:
    print(f"Revisa: {e}")
```
