## Topic 1: Definir y llamar funciones básicas

Una función es un bloque de código reutilizable que realiza una tarea específica.

Sintaxis básica:
```python
def nombre_funcion():
    # código
    print("Hola")
```

Para llamarla: `nombre_funcion()`

Las funciones ayudan a evitar repetir código (principio DRY: Don't Repeat Yourself).

````python
# Demostración: Función sin parámetros
def saludar():
    print("¡Bienvenido al módulo de funciones!")

saludar()
saludar()  # Se puede llamar varias veces
````

## Topic 2: Parámetros y valores por defecto

Las funciones pueden recibir datos (parámetros).

- Posicionales: orden importa
- Por nombre (keyword): especificas el nombre del parámetro
- Valores por defecto: si no se pasa argumento, usa el valor definido

Orden recomendado: parámetros obligatorios → por defecto → *args → **kwargs

````python
# Demostración: Parámetros y defaults
def saludar_persona(nombre, saludo="Hola"):
    print(f"{saludo}, {nombre}!")

saludar_persona("Carlos")                  # Usa default
saludar_persona("María", "Buenos días")    # Sobrescribe default
saludar_persona(saludo="¡Buenas tardes!", nombre="Ana")  # Por nombre
````

## Topic 3: Return y ámbito (scope)

`return` envía un valor de vuelta y termina la función.

Scope:
- Variables dentro de la función → **locales** (solo existen dentro)
- Variables fuera → **globales** (accesibles, pero para modificar usar `global`)

Buena práctica: evitar modificar globales; preferir return.

````python
# Demostración: Return y scope
def sumar(a, b):
    resultado = a + b  # variable local
    return resultado

total = sumar(5, 7)
print("Suma:", total)

# print(resultado)  # Error: NameError, no existe fuera de la función
````

## Topic 4: *args y **kwargs (argumentos variables)

- `*args`: recibe cualquier número de argumentos posicionales como tupla
- `**kwargs`: recibe cualquier número de argumentos por nombre como diccionario

Muy útiles para funciones flexibles (sumar varios números, formatear texto, etc.).

````python
# Demostración: args y kwargs
def sumar_varios(*args):
    return sum(args)

print(sumar_varios(1, 2, 3, 4))      # 10
print(sumar_varios(10))              # 10

def describir_persona(**kwargs):
    for clave, valor in kwargs.items():
        print(f"{clave}: {valor}")

describir_persona(nombre="Luis", edad=35, ciudad="Bogotá")
````

## Topic 5: Funciones lambda (anónimas)

Funciones cortas de una sola expresión, sin nombre.

Útiles con `map()`, `filter()`, `sorted()`, o cuando se pasa una función como argumento.

````python
# Demostración: Lambda
cuadrado = lambda x: x ** 2
print(cuadrado(5))  # 25

numeros = [1, 2, 3, 4, 5]
cuadrados = list(map(lambda x: x**2, numeros))
print("Cuadrados:", cuadrados)
````

### Reto 2: Función con parámetros por defecto y return

Crea una función que calcule área de rectángulo (largo, ancho=10 por defecto). Usa el resultado anterior si aplica. Guarda en `resultado_suma` (simulando reutilización).

````python
# Fallback si no hay mensaje
try:
    mensaje
except NameError:
    mensaje = "¡Hola, Estudiante!"

# Tu código aquí
def area_rectangulo(largo, ancho=10):
    return largo * ancho

resultado_suma = area_rectangulo(8) + area_rectangulo(5, 12)
````

### Reto Final: Función flexible con *args y **kwargs

Crea una función que calcule promedio de números (*args) y agregue info extra (**kwargs). Guarda en `promedio`.

````python
# Fallbacks
try:
    resultado_suma
except NameError:
    resultado_suma = 140

# Tu código aquí
def promedio_info(*args, **kwargs):
    if not args:
        return 0
    prom = sum(args) / len(args)
    print("Info extra:", kwargs)
    return prom

promedio = promedio_info(4, 8, 12, 16, unidad="cm", contexto="muestra")
````

````python
# Validación simple
try:
    assert promedio == 10, "Promedio esperado: (4+8+12+16)/4 = 10"
    print("¡Éxito! Listo para el proyecto.")
except AssertionError as e:
    print(f"Revisa: {e}")
````

#@markdown Pista: sum(args) / len(args) ; print(kwargs)

## Proyecto: Calculadora avanzada modular

Crea varias funciones: suma, resta, multiplicación, división (con manejo de división por cero), y una función principal que reciba operación y números (*args). Usa return y validaciones.

````python
# Tu código aquí - Proyecto: Calculadora avanzada
def operar(operacion, *numeros):
    if not numeros:
        return 0
    if operacion == "suma":
        return sum(numeros)
    elif operacion == "resta":
        return numeros[0] - sum(numeros[1:])
    elif operacion == "multiplicacion":
        resultado = 1
        for n in numeros:
            resultado *= n
        return resultado
    elif operacion == "division":
        if 0 in numeros[1:]:
            return "Error: división por cero"
        resultado = numeros[0]
        for n in numeros[1:]:
            resultado /= n
        return resultado
    else:
        return "Operación no soportada"

# Pruebas
print(operar("suma", 5, 3, 2))          # 10
print(operar("division", 100, 4, 5))    # 5.0
````

````python
# Validación simple del proyecto
try:
    assert operar("suma", 1, 2, 3) == 6, "Suma debería dar 6"
    print("Proyecto completado. ¡Excelente!")
except AssertionError as e:
    print(f"Revisa: {e}")
````
