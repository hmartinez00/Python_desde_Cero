## Topic 1: Repaso integrador – Mini-proyecto: Analizador de texto

Combinamos: archivos, strings, diccionarios, funciones, excepciones.

```python
# Ejemplo integrador: Analizador de texto
def contar_palabras(texto):
    palabras = texto.lower().split()
    conteo = {}
    for palabra in palabras:
        conteo[palabra] = conteo.get(palabra, 0) + 1
    return conteo

try:
    with open('texto_ejemplo.txt', 'w') as f:
        f.write("Python es genial. Python es fácil. Aprende Python hoy!")
    with open('texto_ejemplo.txt', 'r') as f:
        texto = f.read()
    resultado = contar_palabras(texto)
    print(resultado)
except FileNotFoundError:
    print("Archivo no encontrado.")
```

## Topic 2: Decoradores básicos

Un decorador es una función que recibe otra función y extiende su comportamiento.

```python
# Decorador simple: medir tiempo
import time

def medir_tiempo(func):
    def wrapper(*args, **kwargs):
        inicio = time.time()
        resultado = func(*args, **kwargs)
        fin = time.time()
        print(f"{func.__name__} tardó {fin - inicio:.4f} segundos")
        return resultado
    return wrapper

@medir_tiempo
def operacion_lenta():
    time.sleep(1.5)
    return "¡Hecho!"

print(operacion_lenta())
```

## Topic 3: Programación asíncrona básica (asyncio intro)

Útil para tareas I/O-bound (APIs, archivos, redes) sin bloquear el programa.

```python
import asyncio

async def tarea_espera(nombre, segundos):
    print(f"{nombre} iniciado...")
    await asyncio.sleep(segundos)
    print(f"{nombre} terminado!")

async def main():
    await asyncio.gather(
        tarea_espera("Tarea A", 2),
        tarea_espera("Tarea B", 1.5),
        tarea_espera("Tarea C", 3)
    )

asyncio.run(main())
```

### Reto 1: Decorador + función reutilizable

Crea un decorador que registre llamadas y úsalo en una función.

```python
# Tu código aquí
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"Llamando {func.__name__} con args={args}, kwargs={kwargs}")
        resultado = func(*args, **kwargs)
        print(f"Resultado: {resultado}")
        return resultado
    return wrapper

@logger
def multiplicar(a, b):
    return a * b

resultado_api = multiplicar(6, 7)
```

```python
# Validación suave
try:
    assert resultado_api == 42, "multiplicar(6,7) debería dar 42"
    print("¡Bien! Sigue al siguiente reto.")
except AssertionError as e:
    print(f"Intenta de nuevo: {e}")
```

### Reto 2: Async + requests (simulado)

Función async que simula múltiples peticiones (o usa una real).

```python
# Tu código aquí
import asyncio
import random

async def fetch_simulado(url):
    await asyncio.sleep(random.uniform(0.5, 2.0))
    return f"Datos de {url}: OK"

async def main_async():
    urls = ["api1", "api2", "api3"]
    tareas = [fetch_simulado(u) for u in urls]
    resultados = await asyncio.gather(*tareas)
    return resultados

resultados_async = asyncio.run(main_async())
print(resultados_async)
```

```python
# Validación suave
try:
    assert len(resultados_async) == 3, "Deberías tener 3 resultados"
    print("¡Bien! Sigue al reto final.")
except AssertionError as e:
    print(f"Intenta de nuevo: {e}")
```

### Reto Final: Testing + integración

Escribe tests para una función que combines de retos anteriores.

```python
# Tu código aquí
def procesar_datos(*args):
    if not args:
        return 0
    return sum(args) / len(args)

import unittest

class TestProcesar(unittest.TestCase):
    def test_promedio_normal(self):
        self.assertAlmostEqual(procesar_datos(10, 20, 30), 20)

    def test_vacio(self):
        self.assertEqual(procesar_datos(), 0)

    def test_negativos(self):
        self.assertEqual(procesar_datos(-5, 5), 0)

unittest.main(argv=[''], verbosity=2, exit=False)
```

Ejemplo: Gestor de tareas con pronóstico del tiempo para cada día.

```python
# Tu código aquí - Proyecto Capstone (versión simplificada)
import json
import requests
from datetime import datetime

class Tarea:
    def __init__(self, descripcion, fecha):
        self.descripcion = descripcion
        self.fecha = fecha
        self.completada = False

    def __str__(self):
        estado = "✓" if self.completada else " "
        return f"[{estado}] {self.fecha} - {self.descripcion}"

class GestorTareas:
    def __init__(self, archivo="tareas.json"):
        self.archivo = archivo
        self.tareas = self.cargar_tareas()

    def cargar_tareas(self):
        try:
            with open(self.archivo, 'r') as f:
                datos = json.load(f)
                return [Tarea(d["desc"], d["fecha"]) for d in datos]
        except (FileNotFoundError, json.JSONDecodeError):
            return []

    def guardar_tareas(self):
        datos = [{"desc": t.descripcion, "fecha": t.fecha} for t in self.tareas]
        with open(self.archivo, 'w') as f:
            json.dump(datos, f, indent=2)

    def agregar(self, descripcion, fecha):
        self.tareas.append(Tarea(descripcion, fecha))
        self.guardar_tareas()

    def mostrar(self):
        for t in self.tareas:
            print(t)

# Prueba
gestor = GestorTareas()
gestor.agregar("Reunión equipo", datetime.now().strftime("%Y-%m-%d"))
gestor.mostrar()
```
