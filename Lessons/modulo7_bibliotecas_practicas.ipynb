### Reto 1: Uso de random y math

Genera 10 números aleatorios entre 1 y 100, calcula media y desviación. Guarda media en `media_numeros`.

````python
# Tu código aquí
import random
import statistics

numeros = [random.randint(1, 100) for _ in range(10)]
media_numeros = statistics.mean(numeros)
desviacion = statistics.stdev(numeros)

print("Números:", numeros)
print("Media:", media_numeros)
print("Desviación:", desviacion)
````

````python
# Validación suave
try:
    assert 1 <= media_numeros <= 100, "La media debe estar entre 1 y 100"
    print("¡Bien! Sigue al siguiente reto.")
except AssertionError as e:
    print(f"Intenta de nuevo: {e}")
````

### Reto 2: Petición HTTP con requests

Haz una petición a una API pública y guarda un valor en `temperatura`.

````python
# Tu código aquí
import requests

url = "https://api.open-meteo.com/v1/forecast?latitude=40.4165&longitude=-3.70256&current=temperature_2m"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    temperatura = data["current"]["temperature_2m"]
    print(f"Temperatura actual en Madrid: {temperatura} °C")
else:
    temperatura = None
    print("Error en la API")
````

## Proyecto: Mini script de consulta climática + guardado

Crea un script que:
- Use requests para obtener clima de varias ciudades
- Almacene en un DataFrame con pandas
- Guarde el resultado en CSV
- Incluya manejo de errores y comentarios PEP 8

````python
# Tu código aquí - Proyecto: Consulta climática múltiple
import requests
import pandas as pd
from datetime import datetime

def obtener_temperatura(ciudad, lat, lon):
    """Obtiene temperatura actual desde Open-Meteo API."""
    url = f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&current=temperature_2m"
    try:
        r = requests.get(url, timeout=10)
        r.raise_for_status()
        return r.json()["current"]["temperature_2m"]
    except Exception as e:
        print(f"Error en {ciudad}: {e}")
        return None

# Ciudades de ejemplo
ciudades = [
    ("Madrid", 40.4165, -3.70256),
    ("Barcelona", 41.3851, 2.1734),
    ("Sevilla", 37.3891, -5.9845),
    ("Valencia", 39.4699, -0.3763)
]

datos = []
for ciudad, lat, lon in ciudades:
    temp = obtener_temperatura(ciudad, lat, lon)
    if temp is not None:
        datos.append({"Ciudad": ciudad, "Temperatura": temp, "Fecha": datetime.now().strftime("%Y-%m-%d %H:%M")})

if datos:
    df = pd.DataFrame(datos)
    print(df)
    df.to_csv("clima_ciudades.csv", index=False)
    print("\nGuardado en clima_ciudades.csv")
else:
    print("No se obtuvieron datos.")
````

````python
# Validación simple del proyecto
import os
try:
    assert os.path.exists("clima_ciudades.csv"), "No se generó el archivo CSV"
    print("Proyecto completado. ¡Muy bien!")
except AssertionError as e:
    print(f"Revisa: {e}")
````
