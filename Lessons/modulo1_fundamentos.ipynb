## Proyecto: Tip Calculator (calcula propina basada en input).

Crea un programa que pida el monto de la cuenta, calcule una propina (e.g., 15%) y muestre el total. Usa input(), operadores y f-strings. Integra conceptos de retos previos.

```python
# Tu código aquí
# Simulación de input (en ejecución real reemplaza con input())
cuenta = 100.0  # float(input("Monto de la cuenta: "))
propina_porcentaje = 0.15
propina = cuenta * propina_porcentaje
total_con_propina = cuenta + propina
print(f"Cuenta: ${cuenta:.2f} | Propina: ${propina:.2f} | Total: ${total_con_propina:.2f}")
```
