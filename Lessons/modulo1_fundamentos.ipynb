## Tema 2: Hola mundo, variables, tipos de datos (int, float, str, bool).

El clásico 'Hola Mundo' es el primer programa que se escribe en cualquier lenguaje. En Python, es simple: usa print() para mostrar texto. Las variables almacenan datos y no necesitan declarar tipo previo, Python lo infiere automáticamente.

### ¿Qué son los tipos de datos? <a id="t2-tipos"></a>

Un **tipo de dato** es la clasificación que Python asigna a un valor para saber cómo debe interpretarlo y qué operaciones se pueden hacer con él. Python es un lenguaje **dinámicamente tipado**, lo que significa que no tienes que especificar el tipo cuando creas una variable; Python lo detecta automáticamente según el valor que asignes.


### Tipos de datos básicos en Python

A continuación se muestran los cuatro tipos más comunes que usarás al comenzar:

| Tipo | Nombre oficial | Descripción | Ejemplos | Función para verificar |
|------|---|---|---|---|
| **int** | Entero (*integer*) | Números enteros, sin decimales. Pueden ser positivos, negativos o cero. | `5`, `-10`, `0`, `999` | `type(42)` → `<class 'int'>` |
| **float** | Flotante (*floating-point*) | Números con decimales. Se usan para valores fraccionarios o muy grandes/pequeños. | `3.14`, `-2.5`, `0.0`, `1e-3` (notación científica) | `type(3.14)` → `<class 'float'>` |
| **str** | Cadena de texto (*string*) | Texto encerrado en comillas simples `'...'` o dobles `"..."`. Puede incluir espacios, números y símbolos. | `"Hola"`, `'Python'`, `"123"`, `""` (vacío) | `type("texto")` → `<class 'str'>` |
| **bool** | Booleano (*boolean*) | Representa verdadero o falso. Solo tiene dos valores: `True` (verdadero) o `False` (falso). Se usa principalmente en condiciones. | `True`, `False` | `type(True)` → `<class 'bool'>` |

> **Nota:** En Python 3, `True` y `False` deben escribirse con mayúscula inicial.

### Ejemplos y diferencias importantes

- **int vs float:** `5` es un entero, `5.0` es un flotante. Aunque parecen iguales, en memoria se almacenan de forma distinta.
- **"123" vs 123:** La cadena `"123"` es texto (no puedes hacer operaciones matemáticas), mientras que el número `123` es un entero (sí puedes sumar, restar, etc.).

### Conversión entre tipos

Puedes convertir un valor de un tipo a otro usando las funciones `int()`, `float()`, `str()`:
```python
num_texto = "5"
num_entero = int(num_texto)  # Convierte "5" a 5
num_flotante = float(num_texto)  # Convierte "5" a 5.0
resultado = str(num_entero)  # Convierte 5 a "5"
```

⚠️ **Cuidado:** si intentas convertir texto que no es un número, Python lanza un error (`ValueError`).

**Más información:** Puedes consultar la [documentación oficial de tipos de datos en Python](https://docs.python.org/es/3/tutorial/introduction.html#numbers) para una referencia más detallada.

### 3.1 Operadores matemáticos

Los **operadores matemáticos** permiten realizar cálculos con números. Python ofrece los siguientes:

| Operador | Nombre | Descripción | Ejemplo | Resultado |
|---|---|---|---|---|
| `+` | Suma | Suma dos números | `5 + 3` | `8` |
| `-` | Resta | Resta el segundo número del primero | `10 - 4` | `6` |
| `*` | Multiplicación | Multiplica dos números | `6 * 7` | `42` |
| `/` | División | Divide y devuelve un número decimal | `10 / 3` | `3.333...` |
| `//` | División entera | Divide y devuelve solo la parte entera | `10 // 3` | `3` |
| `%` | Módulo (residuo) | Devuelve el residuo de una división | `10 % 3` | `1` |
| `**` | Exponenciación (potencia) | Eleva un número a una potencia | `2 ** 3` | `8` |

#### Consejos importantes:
- **División (`/`) vs División entera (`//`):** `/` siempre devuelve un `float`, mientras que `//` devuelve un `int`.
- **Módulo (`%`):** Es útil para saber si un número es par o impar: `num % 2` da `0` si es par, `1` si es impar.
- **Orden de operaciones:** Python respeta el orden matemático (potencia antes que multiplicación, multiplicación antes que suma, etc.). Usa paréntesis para cambiar el orden: `(2 + 3) * 4` da `20`, no `14`.

---

### 3.2 Input y Output (entrada y salida)

#### **Output: `print()`**

`print()` es una función que muestra texto o valores en pantalla.

```python
print("Hola, Mundo!")  # Texto
print(42)              # Número
print(3.14)            # Decimal
print("El número es:", 42)  # Múltiples valores separados por comas
```

Puedes imprimir múltiples valores separándolos con comas; Python agregará espacios automáticamente.

#### **Input: `input()`**

`input()` es una función que pausa el programa y espera que el usuario escriba algo y presione Enter. Devuelve lo que escribió el usuario, **siempre como texto (`str`)**.

```python
nombre = input("¿Cuál es tu nombre? ")
print(nombre)  # Imprime lo que escribió el usuario
```

⚠️ **Importante:** `input()` siempre devuelve un **string**, incluso si escribes números. Si necesitas un número, debes **convertirlo**:

```python
edad_texto = input("¿Cuántos años tienes? ")  # Usuario escribe: 25
edad = int(edad_texto)  # Convierte "25" a 25 (número)
print(f"El año que viene tendrás {edad + 1}")
```

Si el usuario escribe algo que no se puede convertir, Python lanza un `ValueError`.

---

### 3.3 Strings (cadenas de texto) y manipulación

#### **¿Qué es un string?**

Un **string** (cadena) es una secuencia de caracteres (letras, números, espacios, símbolos) encerrada en comillas simples `'...'` o dobles `"..."`.

```python
texto1 = "Hola"
texto2 = 'Python'
texto3 = '¡Bienvenido!'
```

#### **Concatenación: unir strings**

Puedes unir strings con el operador `+`:

```python
nombre = "Juan"
saludo = "Hola, " + nombre
print(saludo)  # Salida: Hola, Juan
```

#### **F-strings: strings formateados**

Un **f-string** (string formateado) es una forma elegante de insertar variables dentro de un texto. Se escribe con una `f` antes de las comillas y usa `{}` para las variables:

```python
nombre = "María"
edad = 25
print(f"Me llamo {nombre} y tengo {edad} años.")
# Salida: Me llamo María y tengo 25 años.
```

También puedes hacer cálculos dentro de los `{}`:

```python
numero = 5
print(f"El doble de {numero} es {numero * 2}")
# Salida: El doble de 5 es 10
```

#### **Métodos útiles para manipular strings**

| Método | Descripción | Ejemplo | Resultado |
|---|---|---|---|
| `len()` | Devuelve la cantidad de caracteres | `len("Hola")` | `4` |
| `.upper()` | Convierte a mayúsculas | `"python".upper()` | `"PYTHON"` |
| `.lower()` | Convierte a minúsculas | `"PYTHON".lower()` | `"python"` |
| `.strip()` | Elimina espacios al inicio y final | `" Hola ".strip()` | `"Hola"` |
| `.replace()` | Reemplaza texto | `"Hola".replace("a", "x")` | `"Holx"` |
| `.startswith()` | Verifica si comienza con... | `"Hola".startswith("Ho")` | `True` |
| `.endswith()` | Verifica si termina con... | `"Hola".endswith("la")` | `True` |

#### **Slicing (extraer partes de un string)**

Puedes acceder a caracteres o partes de un string usando `[]` con índices:

```python
texto = "Python"
print(texto[0])    # "P" (primer carácter)
print(texto[2])    # "t" (tercer carácter)
print(texto[-1])   # "n" (último carácter)
print(texto[0:3])  # "Pyt" (caracteres del 0 al 2)
print(texto[:3])   # "Pyt" (desde el inicio hasta el 2)
print(texto[3:])   # "hon" (desde el 3 hasta el final)
```

> **Nota:** Los índices comienzan en `0`, no en `1`.

### Solución Parcial - Proyecto: Tip Calculator

Si te atascas en el proyecto, aquí está la estructura paso a paso:

```python
# Paso 1: Obtener la entrada del usuario
cuenta = float(input("Ingresa el monto de la cuenta: "))

# Paso 2: Calcular la propina (15%)
propina_porcentaje = 0.15
propina = cuenta * propina_porcentaje

# Paso 3: Calcular el total
total_con_propina = cuenta + propina

# Paso 4: Mostrar los resultados con formato
print(f"Cuenta: ${cuenta:.2f} | Propina (15%): ${propina:.2f} | Total: ${total_con_propina:.2f}")
```

**Desglose del formato `.2f`:**
- `:.2f` significa "muestra este número como decimal con exactamente 2 dígitos después del punto"
- `$` agrega el símbolo de dólar (puedes cambiar a tu moneda local)
