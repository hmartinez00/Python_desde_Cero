## Topic 1: Listas – Creación, indexing, slicing y métodos comunes

Las listas son colecciones ordenadas y mutables. Se crean con corchetes `[]`.

- Indexing: `lista[0]` (primer elemento), `lista[-1]` (último).
- Slicing: `lista[1:4]` (desde índice 1 hasta antes del 4).
- Métodos clave: `append()`, `extend()`, `insert()`, `remove()`, `pop()`, `sort()`, `reverse()`, `len()`.

Las listas son muy versátiles y se usan para almacenar secuencias de cualquier tipo de dato.

```python
# Demostración: Listas básicas
frutas = ["manzana", "banana", "cereza", "durazno"]

print("Primera:", frutas[0])
print("Última:", frutas[-1])
print("Slicing:", frutas[1:3])

frutas.append("kiwi")
frutas.sort()
print("Lista ordenada:", frutas)
```

## Topic 2: Tuplas – Inmutables y usos comunes

Las tuplas son similares a listas pero **inmutables** (no se pueden modificar después de creadas). Se definen con paréntesis `()` o sin ellos.

Usos típicos:
- Devolver múltiples valores de una función.
- Claves en diccionarios (porque son hashables).
- Datos que no deben cambiar (coordenadas, constantes).

```python
# Demostración: Tuplas
coordenadas = (10.5, 20.3)
print("Latitud:", coordenadas[0])

# Desempaquetado
x, y = coordenadas
print(f"x = {x}, y = {y}")

# Error si intentas modificar
# coordenadas[0] = 15  # TypeError: 'tuple' object does not support item assignment
```

## Topic 3: Sets – Elementos únicos y operaciones de conjuntos

Los sets son colecciones **no ordenadas** y **sin duplicados**. Se crean con llaves `{}` o `set()`.

Operaciones útiles:
- Unión: `|` o `.union()`
- Intersección: `&` o `.intersection()`
- Diferencia: `-` o `.difference()`
- Métodos: `add()`, `remove()`, `discard()`, `clear()`.

```python
# Demostración: Sets
colores_a = {"rojo", "verde", "azul"}
colores_b = {"verde", "amarillo", "azul"}

print("Unión:", colores_a | colores_b)
print("Intersección:", colores_a & colores_b)
print("Diferencia:", colores_a - colores_b)

colores_a.add("morado")
print("Set actualizado:", colores_a)
```

## Topic 4: Diccionarios – Clave-valor, nesting y métodos

Los diccionarios almacenan pares clave-valor. Las claves deben ser inmutables (strings, números, tuplas).

- Acceso: `diccionario["clave"]` o `.get("clave", default)`
- Métodos: `keys()`, `values()`, `items()`, `update()`, `pop()`
- Nesting: diccionarios dentro de diccionarios o listas dentro de diccionarios.

```python
# Demostración: Diccionarios
persona = {
    "nombre": "Ana",
    "edad": 28,
    "habilidades": ["Python", "SQL", "Excel"],
    "direccion": {"calle": "Av. Principal", "ciudad": "Madrid"}
}

print("Nombre:", persona["nombre"])
print("Ciudad:", persona["direccion"]["ciudad"])

persona["edad"] = 29
persona["email"] = "ana@example.com"
print("Diccionario actualizado:", persona)
```

## Proyecto: Generador de contraseñas seguras

Crea un programa que genere una contraseña aleatoria usando listas de caracteres, sets para evitar duplicados y un diccionario para almacenar contraseñas generadas con su categoría (fuerte/media/débil).

```python
# Tu código aquí - Proyecto: Generador de contraseñas
import random

letras = list("abcdefghijklmnopqrstuvwxyz")
mayus = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
numeros = list("0123456789")
simbolos = list("!@#$%^&*()")

todos = letras + mayus + numeros + simbolos
random.shuffle(todos)

longitud = 12
contrasena = "".join(random.sample(todos, longitud))

# Diccionario para clasificar
contrasenas = {
    contrasena: "fuerte" if any(c in mayus+simbolos for c in contrasena) else "media"
}

print("Contraseña generada:", contrasena)
print("Clasificación:", contrasenas[contrasena])
```

```python
# Validación simple del proyecto
try:
    assert len(contrasena) == 12 and isinstance(contrasenas, dict), "Contraseña debe tener 12 caracteres y diccionario de clasificación"
    print("Proyecto completado. ¡Excelente!")
except AssertionError as e:
    print(f"Revisa: {e}")
```
