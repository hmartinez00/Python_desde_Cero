## Glosario rápido
- **Condicional**: estructura que ejecuta código si se cumple una condición (if/else).
- **Operador lógico**: conectores booleanos (and/or/not) para combinar condiciones.
- **Bucle**: repite un bloque de código (for o while).
- **range()**: genera secuencias numéricas (útil en for).
- **break**: sale del bucle inmediatamente.
- **continue**: salta a la siguiente iteración.
- **dummy**: valor temporal que evita errores cuando falta un dato intermedio.

## Topic 1: Estructuras condicionales - if / elif / else

Las condicionales permiten que tu programa tome decisiones según ciertas condiciones. La sintaxis básica es:

```python
if condición:
    # código si verdadero
elif otra_condición:
    # código si la anterior falsa y esta verdadera
else:
    # código si ninguna anterior verdadera
```

La indentación (4 espacios o tab) es obligatoria en Python. Las condiciones suelen usar operadores de comparación: ==, !=, >, <, >=, <=.

### Tabla rápida de condicionales

| Caso | Sintaxis | Descripción |
|---|---|---|
| if | `if condición:` | Ejecuta bloque cuando condición es True |
| elif | `elif otra_condición:` | Se evalúa si if anterior fue False |
| else | `else:` | Se ejecuta si ningún if/elif fue True |

Ejemplo común: validar edad para acceso a contenido.

````python
# Demostración: Condicional simple
edad = 17

if edad >= 18:
    print("Eres mayor de edad. Puedes entrar.")
else:
    print("Eres menor de edad. Acceso restringido.")
````

<a id="topic2"></a>
## Topic 2: Operadores lógicos (and, or, not)

Permiten combinar varias condiciones. Útiles para reglas más complejas, como permisos o filtros.

### Tabla operadores lógicos

| Operador | Significado | Ejemplo |
|---|---|---|
| `and` | True si todas las condiciones son True | `edad >=18 and tiene_licencia` |
| `or` | True si al menos una condición es True | `estado == 'ok' or estado == 'en_proceso'` |
| `not` | Invierte el valor booleano | `not es_vacio` |

````python
# Demostración: Operadores lógicos
edad = 20
tiene_licencia = True
alcohol = 0.0

if edad >= 18 and tiene_licencia and alcohol == 0.0:
    print("Puedes conducir.")
else:
    print("No puedes conducir en este momento.")
````

## Topic 3: Bucles - for y while

**for**: ideal para iterar sobre secuencias conocidas (listas, range, strings).
**while**: repite mientras una condición sea verdadera (cuidado con bucles infinitos).

### Comparación de bucles

| Bucle | Cuándo usar | Ejemplo |
|---|---|---|
| `for` | Secuencia de elementos conocida | `for elem in lista:` |
| `while` | Condición indeterminada | `while contador < 5:` |

La función `range(inicio, fin, paso)` genera secuencias numéricas.

Control: `break` (salir del bucle), `continue` (saltar a la siguiente iteración).

````python
# Demostración: for con range
for i in range(1, 6):
    print(f"Iteración {i}")

# while con break
contador = 0
while True:
    contador += 1
    if contador > 5:
        break
    print(f"Contador: {contador}")
````

### Reto 1: Clasificación de número (condicional)

Pide un número (simulado) y clasifícalo como positivo, negativo o cero. Guarda el resultado en `clasificacion`.

````python
# Tu código aquí
numero = 7  # Cambia o usa input() en ejecución real

if numero > 0:
    clasificacion = "positivo"
elif numero < 0:
    clasificacion = "negativo"
else:
    clasificacion = "cero"
````

````python
# Validación suave
try:
    assert clasificacion in ["positivo", "negativo", "cero"], "Debes clasificar correctamente."
    print("¡Bien! Sigue al siguiente reto.")
except AssertionError as e:
    print(f"Intenta de nuevo: {e}")
````

#### 💡 Pista para el Reto 1

### Reto 2: Contar pares e impares en un rango (for + condicional)

Usa range(1, 21) y cuenta cuántos números pares hay. Guarda en `numeros_pares`.
Si no tienes `clasificacion`, define un dummy.

````python
# Fallback dummy (opcional)
try:
    clasificacion
except NameError:
    clasificacion = "positivo"

# Tu código aquí
numeros_pares = 0
for num in range(1, 21):
    if num % 2 == 0:
        numeros_pares += 1
````

````python
# Validación suave
try:
    assert numeros_pares == 10, f"Deberían haber 10 pares en 1-20. Tienes {numeros_pares}"
    print("¡Bien! Sigue al reto final.")
except AssertionError as e:
    print(f"Intenta de nuevo: {e}")
````

#### 💡 Pista para el Reto 2
- Recorre `range(1, 21)` con `for`.
- Si `num % 2 == 0` es True, incrementa `numeros_pares`.
- Comprueba con `assert numeros_pares == 10`.

Ejemplo:
```python
numeros_pares = 0
for num in range(1, 21):
    if num % 2 == 0:
        numeros_pares += 1
```

### Reto Final: Validación combinada (while + condicional + operadores lógicos)

Simula un login: pide contraseña hasta 3 intentos. Usa while y operadores lógicos. Guarda en `es_valido` si acertó.

````python
# Fallbacks
try:
    numeros_pares
except NameError:
    numeros_pares = 10

# Tu código aquí
intentos = 0
contrasena_correcta = "python123"
es_valido = False

while intentos < 3:
    # contraseña = input("Ingresa contraseña: ")
    contraseña = "python123"  # Simulación para prueba
    if contraseña == contrasena_correcta:
        es_valido = True
        break
    intentos += 1

print(f"Acceso {'concedido' if es_valido else 'denegado'} después de {intentos} intentos.")
````

````python
# Validación simple
try:
    assert es_valido == True, "El sistema debería validar correctamente en esta simulación."
    print("¡Éxito! Listo para el proyecto.")
except AssertionError as e:
    print(f"Revisa: {e}")
````

#### 💡 Pista para el Reto Final
- `intentos` empieza en 0 y aumenta en cada ciclo.
- `while intentos < 3:` verifica el límite de intentos.
- Si `contraseña == contrasena_correcta`, establece `es_valido = True` y `break`.
- Si falla, incrementa `intentos`.

Ejemplo:
```python
while intentos < 3:
    contraseña = "python123"
    if contraseña == contrasena_correcta:
        es_valido = True
        break
    intentos += 1
```
